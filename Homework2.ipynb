## Tính diện tích, chu vi hình chữ nhật

```python
chieu_dai, chieu_rong = map(float, input("Nhập chiều dài, chiều rộng: ").split())
dien_tich = chieu_dai * chieu_rong
chu_vi = (chieu_rong + chieu_dai)*2
print(f"Diện tích là {dien_tich}, chu vi là {chu_vi}")
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
import math
a, b, c = map(int, input("Nhập 3 cạnh của tam giác").split())
if a + b > c and a + c > b and b + c > a:
    if a == b and a == c and c == b:
        print("Đây là tam giác đều")
    elif a == b or a == c or b == c:
        print("Đây là tam giác cân")
    elif a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == b**2 + a**2:
        print("Đây là tam giác vuông")
    else:
        print("Đây là tam giác thường")
    p = (a + b + c)/2
    dien_tich = math.sqrt(p*(p-a)*(p-b)*(p-c))
    chu_vi = a + b + c
    print(f"Diện tích tam giác là {dien_tich}, chu vi là {chu_vi}")
else:
    print("Khong la ba canh cua mot tam giac")

```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a, b, c = map(float, input("Nhập a, b, c: ").split())
denta = b**2 - 4*a*c
if denta >= 0:
    x1 = (-b - math.sqrt(denta))/(2*a)
    x2 = (-b + math.sqrt(denta))/(2*a)
    print(f"Nghiệm đầu là {x1}, nghiệm 2 là {x2}")
else:
    print("Phương trình vô nghiệm")

```

# Bài 8: Giải hệ phương trình tuyến tính: 
# ax + by = m 
# cx + dy = n 

```python
a, b, m, c, d, n = map(float, input("Nhập a, b, m, c, d, n: ").split())
D = (a*d) - (b*c)
Dx = (m*d) - (b*n)
Dy = (a*n) - (c*m)
if D != 0:
    x1 = Dx / D
    x2 = Dy / D
    print(f"Nghiệm của phương trình là {x1} và {x2}")
else:
    if Dx == 0 and Dy == 0:
        print("Phương trình vô số nghiệm")
    elif Dx != 0 or Dy != 0:
        print("Phương trình vô nghiệm")
    
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây. 

```python
thoi_gian = int(input("Nhập số giây: "))
gio = thoi_gian // 3600
phut = ((thoi_gian/3600) - gio) // (1/60)
giay = round(((((thoi_gian/3600) - gio) / (1/60)) - phut) / (1/60))
print(f"{gio} giờ {phut} phút {giay} giây")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không? 

```python
import math
a, b = map(float, input("Nhập tọa độ tâm: ").split())
r = float(input("Nhập bán kính: "))
x, y = map(float, input("Nhập tọa độ điểm A bất kì: ").split())
do_dai = math.sqrt(((x-a)**2) + ((y-b)**2))
if do_dai > r:
    print("Không thuộc đường tròn")
else:
    print("Thuộc đường tròn")
```
