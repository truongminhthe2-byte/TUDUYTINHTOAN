## W3A3 - Lũy thừa của 2
Kiểm tra xem 1 số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem 1 số có phải là lũy thừa của 2 hay không
Gợi ý: Đối với lũy thừa của 2, n & (n-1) = 0

```python
nhap_so = int(input())
if nhap_so > 0 and nhap_so & (nhap_so - 1) == 0:
    print(f"{nhap_so} là lũy thừa của 2")
else:
    print(f"{nhap_so} không phải lũy thừa của 2")

```
