Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường 
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem 
trước cấu trúc lựa chọn: if….else, ma)


```python
# bai 3
def ki_tu(kitu):
    if kitu.isupper():
        print(kitu.lower())
    elif kitu.islower():
        print(kitu.upper())
nhap_ki_tu = input("nhập kí tự: ")
ki_tu(nhap_ki_tu)
```

Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái 
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} 
không phải là kí tự alphabet. 

```python
# bai 4
def chu_cai(chucai):
    if chucai.isalpha():
        print(f"{chucai} là kí tự alphabet")
    else:
        print(f"{chucai} không phải là kí tự alphabet")
nhap = input("Nhập chữ cái: ")
chu_cai(nhap)
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương 
ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 

```python
# bai 5
chu_cai = input("nhập chữ cái in hoa: ")
vt = ord(chu_cai)
if ord("A") < vt <= ord("Z"):
    trc = vt - 1
    print(chr(trc).lower())
elif chu_cai == "A":
    print("Không có chữ cái trước chữ A")




```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, 
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó 
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
# bai 6
import math
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c)/2
    S = math.sqrt(p*(p-a)*(p-b)*(p-c))
    print("Đây là 1 tam giác có diện tích là " + str(round(S, 1)))
else:
    print("Khong phai 3 canh cua tam giac")
```

tính tiền điện tháng này cho hộ biết:  -

Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh

Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh  
Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh  
Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh

```python
# bai 8
ten_chu_ho = input("Ten chu ho: ")
thang_truoc = int(input("Chi so thang truoc: "))
thang_nay = int(input("Chi so thang nay: "))
sodien = thang_nay - thang_truoc
if 0 <= sodien <= 50:
    tienphaitra = (sodien * 1984) * (1+ 0.08)  
elif 51 <= sodien <= 100:
    tienphaitra = (50*1984 + (sodien-50*1)*2050)*(1+0.08)
elif 101 <= sodien <= 200:
    tienphaitra = (50*(1984+2050) + (sodien-50*2)*2380)*(1+0.08)
elif 201 <= sodien <= 300:
    tienphaitra = (50*(1984+2050) +100*2380 + (sodien-50*2-100)*2998)*(1+0.08)
elif 301 <= sodien <= 400:
    tienphaitra = (50*(1984+2050)+100*(2380+2998) + (sodien-50*2-100*2)*3350)*(1+0.08)
else:
    tienphaitra = (50*(1984+2050)+100*(2380+2998+3350) + (sodien-50*2-100*3)*3460)*(1+0.08)
print(f"Ho va ten: {ten_chu_ho}" )
print("Tien phai tra la: " + str(round(tienphaitra)))

    
```
