[ModAndDiv] 
Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tổng, hiệu, tích, phần 
nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập 
phân) theo định dạng đúng của output 

```python
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
Tong = a + b
Hieu = a - b
Tích = a * b
Thuong_nguyen = a // b
Thuong_du = a % b
Thuong_thuc = a / b
print(f"Tổng {a} + {b} là: {Tong}")
print(f"Hiệu là: {Hieu}")
print(f"Chia lấy phần nguyên là: {Thuong_nguyen}")
print(f"Chia lấy phần dư bằng: {Thuong_du}")
print("Chia thực là: " + str(round(Thuong_thuc, 2)))
```

[AverageOfSix] 
Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 
6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó: 
• Các điểm a1, b1, c1 được nhân hệ số 1, 
• Các điểm a2, b2 được nhân hệ số 2, 
• Điểm a3 được nhân hệ số 3. 
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10 
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6 
đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được 
phân tách nhau bởi duy nhất 1 dấu cách. 
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân. 

```python
a1,b1,c1,a2,b2,a3 = map(float, input("Nhập 6 đầu điểm: ").split())
Diem_trung_binh = ((a1+b1+c1)+(a2+b2)*2+(a3)*3)/10
print(f"Điểm trung bình 6 môn học là {Diem_trung_binh:.1f}")
```

[Exponentiation] 
Trong Python, phép toán mũ (luỹ thừa) được viết bằng **. Ví dụ: để tính 2^3 bạn viết 2**3, và kết 
quả là 8. Cơ số và số mũ có thể có kiểu float hoặc int. Trong bài này bạn hãy nhập vào hai số a, b 
(int) cách nhau một dấu cách, tính và in ra kết quả a mũ b.

```python
a, b = map(int, input("Nhập a, b ").split())
mu = a**b
print(f"{a} mũ {b} bằng {mu}")
```

[StringDataType] 
“Chuỗi” trong python là các kí tự được xâu lại với nhau như những hạt trên một chiếc dây chuyền. 
Chính vì thế “chuỗi” còn được gọi là “xâu”. Một chuỗi gồm các kí tự nằm trong dấu nháy đơn '' hoặc 
nháy đôi "”.• Mỗi chuỗi (str) là một chuỗi các kí tự trong bảng mã Unicode kể từ Python 3 trở đi 
(Unicode là một tiêu chuẩn mã hóa kí tự toàn cầu, cho phép biểu diễn hầu hết mọi kí tự trong tất cả 
các ngôn ngữ trên thế giới, bao gồm cả kí hiệu, biểu tượng, emoji,….). Bạn có thể dùng ord() và 
chr() để làm việc với mã Unicode.  
Hãy nhập vào một kí tự chữ cái thường ‘a’…’z’, in ra mã Unicode của nó sau đó chuyển nó thành kí 
tự hoa tương ứng, in kí tự hoa này ra màn hình. 

```python
kitu = input("Nhập kí tự in thường: ")
ma_unicode = ord(kitu)
if kitu.islower():
    print("Kí tự in hoa là: " + kitu.upper())
    print("Mã unicode của kí tự " + kitu + " là: " + str(ma_unicode))
# ord(A) = ord(a) - 32 "Kí tự hoa bằng kí tự thường trừ đi 32 "
```

Viết chương trình chuyển từ giờ và phút sang giây 
Đầu vào gồm hai giá trị chứa số giờ \h và số phút \m:  
Đầu ra: In ra màn hình một dòng duy nhất số giờ và phút được chuyển sang đơn vị giây. 


```python
a, b = map(int, input("Nhập số giờ, số phút: ").split())
c = a*3600 + b*60
print(f"{a} giờ {b} phút chuyển thành {c} giây ")
```

Cho một khối Rubik có độ dài cạnh là n, hãy trả về số lượng miếng dán riêng lẻ cần thiết để 
bao phủ toàn bộ khối lập phương. Ví dụ, khối rubic có độ dài cạnh là 1 cần 6 miếng dán 
riêng lẻ. (Lưu ý: một khối rubic có 6 mặt) 

```python
n = float(input("Độ dài cạnh rubik là: "))
so_mieng_dan = n*n*6
print(f"Số miếng dán cần dùng là {so_mieng_dan}")
```
